```python
import requests

r=requests.get('https://aaai-2022.virtualchair.net/papers.json')
#print(r.json())
json=r.json()
```

```python
import os
import time
#create a "\paper"  dir
path=os.getcwd()
path=path+'\\paper\\'
print(path)
if not os.path.exists("paper"):
    os.makedirs("paper")
print("begin"+time.strftime('%H:%M:%S',time.localtime(time.time())))
```

```python
from urllib.error import URLError
from urllib.request import ProxyHandler,build_opener
#proxy
proxy='127.0.0.1:999'  #you should set your proxy to speed up downloa
proxy_handler=ProxyHandler({
    'http':'http://'+proxy,
    'https':'https://'+proxy
})
opener=build_opener(proxy_handler)

```

```python
# proxy test
try:
    response=opener.open('https://www.baidu.com/')
    print(response.read().decode('utf-8'))
except URLError as e:
    print(e.erason)
```

```python
from urllib.request import urlopen
from urllib.parse import quote

from io import open
import time

set_cluster=[]
error_message="#\r\n"
success_message="#1\r\n"

for item in json:
    try:
        cluster=item.get('cluster')
        title=item.get('title')
        title_temp=title
        #only alphabet
        title = ''.join(filter(lambda i:  i in [' ']  or i.isalpha() or i.isdigit() ,list(title) ))
        print(title)
        
        paper_url=item.get('paper')
        #replace "spacing"  by utf-8 
        paper_url = paper_url.replace(" ", quote(' ', 'utf-8'))
        
        #paper_url = quote(paper_url, 'utf-8')
        print(paper_url)
        
        #--------------------------------------
        r = opener.open(paper_url)  #url
        
        #if there is no this cluster ,create a dir
        if cluster not in set_cluster:
            path_final=path+'\\'+cluster
            set_cluster.append(cluster)

            sub_path="paper\\"+cluster
            if not os.path.exists(sub_path):
                os.makedirs("paper\\"+cluster)
        #----------------------------------------------------
        fo = open(path+'\\'+cluster+'\\'+title+'.pdf','wb')                       
        fo.write(r.read())                             
        fo.close()
        
        my_time=time.strftime('%H:%M:%S',time.localtime(time.time()))+"---"
        message="finished"+"---"+cluster+"---"+title+'---'+paper_url+'\r\n'
        success_message+=my_time+message
        print(my_time+message)
        #print(paper_url)
     

    except Exception as e:
        my_time=time.strftime('%H:%M:%S',time.localtime(time.time()))+"---"
        message="error"+"---"+cluster+"---"+title+'---'+paper_url+'\r\n'
        error_message+=my_time+message
        print(e)
        fo.close()
      

fo = open("error_message.txt",'w')                       
fo.write(error_message)                             
fo.close()

fo = open("success_message.txt",'w')                       
fo.write(success_message)                             
fo.close()
```
